```python
import pandas as pd
import numpy as np

df = pd.read_csv('merged_dataset.csv')

df.head(10)
```

```python
def filterName(row):
    name_tokens = row['Name'].lower().split()
    if('babyfood' in name_tokens):
        return row['Name'], row['Fat'], row['Calcium']

babyfoods = df.apply(filterName, axis=1)

name = list()
fat = list()
calcium = list()

for food in babyfoods:
    if(food != None):
        name.append(food[0])
        fat.append(food[1])
        calcium.append(food[2])
```

```python
from pandas.core.frame import DataFrame


source = DataFrame(
    dict(
        Fat=fat,
        Calcium=calcium,
        Name=name
    )
)
source.to_csv('babyfood.csv')
```

```python
def filterName(row):
    name_tokens = row['Name'].lower().split()
    if('chips' in name_tokens):
        return row['Name'], row['Calories'], row['Fat']

chips = df.apply(filterName, axis=1)

name = list()
calories = list()
fat = list()

for food in chips:
    if(food != None):
        name.append(food[0])
        calories.append(food[1])
        fat.append(food[2])
```

```python
source = DataFrame(
    dict(
        Name=name,
        Fat=fat,
        Calories=calories,
    )
)
source.to_csv('chips.csv')
```
