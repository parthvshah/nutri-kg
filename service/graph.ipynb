```python
import kglab
import pandas as pd
```

```python
namespaces = {
    "flavor": "https://dsci558.org/Flavor",
    "food": "https://dbpedia.org/ontology/Food",
    "nutriinfo": "https://schema.org/NutritionInformation",
    "foodcategory": "https://dsci558.org/FoodCategory/",
    "xsd": "http://www.w3.org/2001/XMLSchema#"
}
```

```python
kg = kglab.KnowledgeGraph(
    name = "NutriKG",
    namespaces = namespaces,
)
```

```python
kg.load_rdf("large_graph.ttl")
```

```python
sparql = """
SELECT ?p ?o ?r
WHERE {
    ?p ?o ?r .
}
"""


df = kg.query_as_df(sparql)
df.head(20)
```

```python
food = "crunch n munch buttery toffee popcorn with peanuts 6 oz by conagra brands"
food = food.replace(" ", "-")

sparql = """
SELECT ?o ?c2 ?c1
WHERE {
    food1:crunch-n-munch-buttery-toffee-popcorn-with-peanuts-6-oz-by-conagra-brands flavor1:flavorOf ?o .
    food1:crunch-n-munch-buttery-toffee-popcorn-with-peanuts-6-oz-by-conagra-brands nutriinfo1:calories ?c1 .
    ?o nutriinfo1:calories ?c2 .
    FILTER(?c2 < ?c1)
}
    ORDER BY ?c2
"""


df = kg.query_as_df(sparql)
df.head(20)
```

```python
import pandas as pd

cats = pd.read_csv('data/Nutritional Data with Categories UTF-8.csv')
no_cats = pd.read_csv('data/Sample - Nutrition Value - Source 1.csv')
```

```python
sub_cats_list = []
cats_list = []

no_cats.reset_index()
for index, row in no_cats.iterrows():
    search = row['Food Name']
    
    sub_cat = cats.loc[cats['Food Name'] == search]
    sub_cat.reset_index(inplace=True)
    cat = cats.loc[cats['Food Name'] == search]
    cat.reset_index(inplace=True)

    print(search, "==", sub_cat['Food Name'][0], "==", cat['Food Name'][0])
    print(sub_cat['Sub-Category'][0], cat['Category'][0])
    print("=============")

    sub_cats_list.append(sub_cat['Sub-Category'][0])
    cats_list.append(cat['Category'][0])
```

```python
print(len(sub_cats_list))
print(len(cats_list))
print(len(no_cats))

no_cats['Sub-Category'] = sub_cats_list
no_cats['Category'] = cats_list

no_cats.head(5)
no_cats.to_csv('data/Sample - Nutrition Value with Categories - Source 1.csv', index=False)
```
